## 3.1. Loading the dataset of Tsunami events into a Pandas dataframe

````python
# 1. Code to load dataset into the dataframe

# To begin with, we import the Pandas dataframe and represent it using pd for ease of usage in writing further code
import pandas as pd 

# We read the data from file 'data/tsevent.csv' using the function 'read_csv' within the Pandas package
if 'google.colab' in str(get_ipython()):
    data = pd.read_csv('https://raw.githubusercontent.com/loftytopping/DEES_programming_course/master/data/tsevent.csv')
else:
    data = pd.read_csv("data/tsevent.csv")
    
#The '.' operator is used to access the attributes belonging to different object classes.
# We preview the first 5 lines of the loaded data using '.head(5)' 
data.head(5)

#### ------------------------------------
````

## 3.2. Geospatial plots of tsunami events
We extract the data from the loaded Pandas dataframe and generate a global plot of locations of tsunami events using Cartopy and Matplotlib.

In order to avoid repetition and for the ease of writing the code, with the aid of Python we define a function fig() to create a canvas of the world map and call this function in the subsequent codes.

````python
# We import Cartopy and Matlpotlib as ccrs and plt respectively.
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# We define the function 'fig()' to create a canvas portraying the world map on which we shall plot our tsunami events.
def fig():
    fig=plt.figure(figsize=(15,15))

    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.stock_img()
    return #This statement allows a function to return the value
````

### 3.2.1 Plotting all global tsunami events on the world map

````python
## Code to select required data from the dataframe and plot it on the map.

# We call the already defined function fig() to create a canvas of the world map.
fig()

#We extract the 'LATITUDE' and 'LONGITUDE' columns and then their .values
Latitude = data['LATITUDE'].values 
Longitude = data['LONGITUDE'].values 

# We plot the extracted data points on the map using blue 'o' symbols  
plt.scatter(Longitude, Latitude,
         color='blue', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
````

####                  Figure 2: Map with geospatial plots of tsunami occurences 

### 3.2.2. Plotting tsunami events occurred after 2000 and not by a questionable earthquake
We use the '&' operator in order to plot the tsunami events satisfying both the conditions stated above. 

````python
# Code to plot tsunamis that have occurred after 2000 and not by a questionable earthquake

# We call the already defined function fig() to create a canvas of the world map.
fig()

# We extract the 'LATITUDE' and 'LONGITUDE' columns and their .values where the 'YEAR'>2000 and the event is 'not caused by a questionable earthquake' and then its .values
Latitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LATITUDE'].values   
Longitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LONGITUDE'].values   

# We plot the extracted data points on the map using black 'o' symbols
plt.scatter(Longitude, Latitude,
         color='black', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
#
````

#### Figure 3: Geospatial plot of tsunamis occured after 2000 by a non-questionable earthquake

### 3.2.3. Comparison of the geospatial plots of tsunamis caused by volcanoes and earthquakes 

````python
# Code to create subplots presenting the volcano and earthquake induced tsunamis. 

# We create a canvas to plot our specific locations on the subplots. 'nrows' and 'ncols' mention the number of rows and columns of the subplots respectively
fig, axs = plt.subplots(nrows=1,ncols=2,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(20,20))


axs[0].stock_img()
axs[1].stock_img()

# We extract the 'LATITUDE' and 'LONGITUDE' columns and their .values for events caused by volcanoes and earthquakes independently. 
Latitude = data[data['CAUSE_CODE']==6]['LATITUDE'].values 
Longitude = data[data['CAUSE_CODE']==6]['LONGITUDE'].values 
Latitude1 = data[data['CAUSE_CODE']==1]['LATITUDE'].values
Longitude1 = data[data['CAUSE_CODE']==1]['LONGITUDE'].values 

# We plot the extracted data points using red and green 'o' symbols respectively.
axs[0].scatter(Longitude, Latitude,
         color='red', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )
axs[1].scatter(Longitude1, Latitude1,
            color='green', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )

plt.show()
````

#### Figure 4(a): Geospatial plot of tsunamis occured due to volcanic eruptions (left)
#### Figure 4(b): Geospatial plot of tsunamis caused by earthquakes (right)


## 3.3. Assessment of intensity of the tsunamis using histograms
Presenting the histograms depicting the intensity of tsunamis induced by volcanoes and earthquakes separately, and comparing them combinedly with those induced by questionable earthquakes to ultimately plot a comprehensive histogram. 

### 3.3.1. Histogram presenting the intensity of tsunamis created by earthquakes 

````python
# Code to plot a histogram depicting the intensity of volcano-induced tsunamis
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[data['CAUSE_CODE']==6].hist(column='SOLOVIEV',ax=axes)

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes.set_ylabel('Probability Frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 5: Histogram of intensities of tsunamis induced by volcanoes 

### 3.3.2. Histogram presenting the intensity of tsunamis created by earthquakes 

````python
# Code to plot a histogram depicting the intensity of earthquake-induced tsunamis 
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[data['CAUSE_CODE']==1].hist(column='SOLOVIEV',ax=axes)

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes.set_ylabel('Probability frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 6: Histogram of intensities of tsunamis caused by earthquakes

### 3.3.3. Comparing the intensity of tsunamis caused by volcanoes and earthquakes with those caused by the questionable earthquakes

````python
#Code to plot and compare histograms of the intensities of tsunamis caused by volcanoes and earthquakes combinedly with the intensities of the tsunamis caused by questionable earthquakes
fig, axes = plt.subplots(nrows=1, ncols=2,
                        figsize=(10,10))

data[data['CAUSE_CODE']==4].hist(column='SOLOVIEV',ax=axes[0])# Plotting the histogram for assessing tsunami intensities caused by volcanoes and earthquakes combinedly
data[data['CAUSE_CODE']==2].hist(column='SOLOVIEV',ax=axes[1])# Plotting the histogram for assessing tsunami intensities caused by questionable earthquakes

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes[0].set_ylabel('Probability frequency') 
axes[0].set_xlabel('Tsunami intensity') 
axes[1].set_ylabel('Probability frequency') 
axes[1].set_xlabel('Tsunami intensity')
````

#### Figure 7(a) Histogram of intensities of tsunamis induced by volcanoes and earthquakes (left) 7(b) Histogram of intensities of tsunamis induced by questionable earthquakes (right)

### 3.3.4. Comprehensive analysis of the intensity of tsunamis caused by volcanoes and earthquakes especially by the non-questionable earthquakes.

````python
# Code to plot a histogram depicting the intensity of tsunamis induced by volcanoes and earthquakes especially the ones induced by non-questionable earthquakes
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[(data['CAUSE_CODE']==4) & (data['CAUSE_CODE']!=2)].hist(column='SOLOVIEV',ax=axes)

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes.set_ylabel('Probability frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 8: Histogram of tsunamis induced by volcanoes and earthquakes especially non-questionable earthquakes

# Results

The plots and figures shown above highlight the significant relationship between the intensity of tsunamis and their causes related to two major geological hazards namely, earthquakes and volcanoes.

Figure 2 shows all the tsunami occurrences from 2100 BC to present while figure 3 specifically highlights those tsunami events that occurred due to unquestionable earthquakes after the year 2000. The stark contrast between figures 2 and 3 indicates that very few tsunami events are caused by unquestionable earthquakes after 2000 which could be due to relatively longer reporting period of tsunamis before 2000 and higher rates of tectonic activity before 2000. 

We can confirm our result by executing the following code which shows that only 248 events out of 2674 events are caused after 2000 by unquestionable earthquakes.


````python
# Finding the number of tsunamis caused after 2000 by a non-questionable earthquake
((data['YEAR']>2000) & (data['CAUSE_CODE']!=2)).value_counts()
````

On comparison of the subplots in figure 4 showing the distribution of the volcano and earthquake derived tsunami intensities, we can comprehend that relatively a greater number of tsunamis occur due to earthquakes than due to volcanoes. We can verify this by executing the following code which shows that more than half of the total events (1903 out of 2674 tsunamis) were caused by earthquakes. 

````python
# Finding the number of tsunamis caused by earthquakes
(data['CAUSE_CODE']==1).value_counts()
````

During the assessment of the intensity of tsunamis, asymmetric distributions are obtained wherein plot 5 shows that the maximum number of tsunamis (16) induced by volcanoes have an intensity between 0.5 to 1 while plot 6 portrays that the maximum number of earthquake-induced tsunamis (approximately 310) have an intensity around 0-1. This means that majority of the tsunamis caused by volcanoes and earthquakes are of low intensity. However, we also observe a higher distribution of earthquake-induced tsunamis having intensities more than 0 as well as the maximum earthquake-derived tsunami intensity between 7.8 to 8.5 (from plot 5) which is greater than the maximum volcano-derived tsunami intensity between 5.4 to 6 (from plot 6) which emphasizes the role of earthquakes in causing highly impactful tsunamis. This can be linked to the larger and more frequent direct impacts of earthquake-induced seismic waves on the sea floor compared to the occasional volcanic eruptions triggered mostly by decompression melting and mantle plume hotspots (Hinga, 2015).

The histogram in figure 7(b) tells us that three tsunami events (two with intensities ranging from 1 to 1.25 and one with an intensity between 2.75 to 3) are caused by questionable earthquakes which can be excluded from the tsunami events data represented in Figure 7(a) to improve the precision of our analysis. Therefore, in the last figure 8, we extend our code and comprehensively depict the intensities of tsunamis caused by volcanoes and earthquakes excluding those caused by questionable earthquakes. We find that there are considerable number of moderate intensity events caused by the combined effect of volcanoes and non-questionable earthquakes indicating that there might be a link between the two geological events which leads to the formation of tsunamis.
