## 1. Loading the dataset of Tsunami events into a Pandas dataframe

````python
# 1. Code to load dataset into the dataframe

# To begin with, we import the Pandas dataframe and represent it using pd for ease of usage in writing further code
import pandas as pd 

# We read the data from file 'data/tsevent.csv' using the function 'read_csv' within the Pandas package
if 'google.colab' in str(get_ipython()):
    data = pd.read_csv('https://raw.githubusercontent.com/loftytopping/DEES_programming_course/master/data/tsevent.csv')
else:
    data = pd.read_csv("data/tsevent.csv")
    
#The '.' operator is used to access the attributes belonging to different object classes.
# We preview the first 5 lines of the loaded data using '.head(5)' 
data.head(5)

#### ------------------------------------
````

## 2. Geospatial plots of tsunami events
Extracting data from the loaded Pandas dataframe and creating a global plot of locations of events using Cartopy and Matplotlib.

````python
# 2. We import Cartopy and Matlpotlib as ccrs and plt respectively.
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# We define a function 'fig()' to create a canvas comprising of the map on which we shall plot our tsunami events.
def fig():
    fig=plt.figure(figsize=(15,15))
    fig = plt.figure(figsize=(15,15))

    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.stock_img()
    return
````

### 2.1. Plotting all global tsunami events on the world map

````python
## 2.1 Code to select required data from the dataframe and plot it on the map.

# We call the already defined function fig() to create a canvas of the world map.
fig()

Latitude = data['LATITUDE'].values # We extract the 'LATITUDE' column and then its .values 
Longitude = data['LONGITUDE'].values # We extract the 'LATITUDE' column and then its .values

# We plot the extracted data points on the map using blue 'o' symbols  
plt.scatter(Longitude, Latitude,
         color='blue', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
````

####                  Figure 1: Map with geospatial plots of Tsunami occurences 

### 2.2. Plotting tsunami events occured after 2000 and not by a questionable earthquake


````python
# 2.2 Code to plot tsunamis that have occurred after 2000 and not by a questionable earthquake

# We call the already defined function fig() to create a canvas of the world map.
fig()

Latitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LATITUDE'].values  # We extract the 'LATITUDE' column where the 'YEAR'>2000 and the event is not caused by a questionable earthquake and then its .values
Longitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LONGITUDE'].values   #We extract the 'LONGITUDE' column where the 'YEAR'>2000 and the event is not caused by a questionable earthquake and then its .values 

# We plot the extracted data points on the map using black 'o' symbols
plt.scatter(Longitude, Latitude,
         color='black', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
#
````

#### Figure 2: Geospatial plot of tsunamis occured after 2000 by a non-questionable earthquake

### 2.3. Plotting the tsunami events along the Pacific Rim also known as the Ring of Fire
The Pacific Rim refers to the most tectonically active region extending between the Aleutian islands 52.891567° N, -170.93575° E and New Zealand -42 degrees N, 174 degrees E. Thus, to visualise the number of events occurring in this region, we plot the tsunami events lying between -40 ° N and 53 ° N.

````python
# We plot the tsunami events along the Pacific Rim which lies within the boundaries marked by-40 ° N and 53 ° N.

# We call the already defined function fig() to create a canvas of the world map.
fig()

# Extract the 'LATITUDE' and 'LONGITUDE' columns and their .values within the above-mentioned boundaries
Latitude = data[(data['LATITUDE']> -40) & (data['LATITUDE']< 53)]['LATITUDE'].values 
Longitude = data[(data['LATITUDE']> -40) & (data['LATITUDE']< 53)]['LONGITUDE'].values   

# We plot the extracted data points using brown 'o' symbols
plt.scatter(Longitude, Latitude,
         color='brown', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
````

#### Figure 3. Tsunamis along the Pacific Rim

### 2.4. Comparison of the geospatial plots of tsunamis caused by volcanoes and earthquakes 

````python
# 2.4 We plotting and analysing the relation between volcano and earthquake induced tsunamis. 

# Creating a canvas to plot our specific locations.
fig, axs = plt.subplots(nrows=1,ncols=2,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(20,20))


axs[0].stock_img()
axs[1].stock_img()

# Extracting the 'LATITUDE' and 'LONGITUDE' columns and their .values for events caused by volcanoes and earthquakes independently. 
Latitude = data[data['CAUSE_CODE']==6]['LATITUDE'].values 
Longitude = data[data['CAUSE_CODE']==6]['LONGITUDE'].values 
Latitude1 = data[data['CAUSE_CODE']==1]['LATITUDE'].values
Longitude1 = data[data['CAUSE_CODE']==1]['LONGITUDE'].values 

# We plot the extracted data points using red and green 'o' symbols respectively.
axs[0].scatter(Longitude, Latitude,
         color='red', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )
axs[1].scatter(Longitude1, Latitude1,
            color='green', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )

plt.show()

#---------------------------------------------
````

#### Figure 4(a): Geospatial plot of tsunamis occured due to volcanic eruptions (left)
#### Figure 4(b): Geospatial plot of tsunamis caused by earthquakes (right)


## 3. Intensity assessing histograms
Presenting the intensity for Tsunamis created by volcanoes, earthquakes, and comparing them with the questionable earthquakes through histograms. According to the National Centers for Environmental Information (NCEI) database, valid values of tsunami intensities range from -5 to 10 and tsunami intensity is defined by Soloviev and Go (1974) as


I = \log _{2} (2^{\frac{1}{2}}\times h)                                                                  (1)

where "h" is the average runup height of the wave. 

### 3.1. Assessing intensity of tsunamis caused by volcanoes

````python
# We plot a histogram depicting the intensity of volcano-induced tsunamis using the following code:

fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[data['CAUSE_CODE']==6].hist(column='SOLOVIEV',ax=axes)

# We label the x axis as 'Tsunami intensity' and the y axis as 'Probability frequency'
axes.set_ylabel('Probability Frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 5: Histogram of intensity of tsunamis induced by volcanoes 

### 3.2. Presenting the intensity for Tsunamis created by earthquakes through histogram 

````python
# We plot a histogram depicting the intensity of earthquake-induced tsunamis using the following code: 
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[data['CAUSE_CODE']==1].hist(column='SOLOVIEV',ax=axes)

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes.set_ylabel('Probability frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 6: Histogram of intensity of tsunamis caused by earthquakes

### 3.3 Assessing the intensity of tsunamis caused by volcanoes and earthquakes especially by the non-questionable earthquakes.

````python
# We plot a histogram depicting the intensity of tsunamis induced by volcanoes and earthquakes especially the ones induced by non-questionable earthquakes
fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True)
data[(data['CAUSE_CODE']==1) & (data['CAUSE_CODE']!=2)].hist(column='SOLOVIEV',ax=axes)

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes.set_ylabel('Probability frequency')
axes.set_xlabel('Tsunami intensity')
````

#### Figure 7: Histogram of tsunamis induced by earthquakes

### 3.4. Comparing the intensity of tsunamis caused by volcanoes and earthquakes with those caused by the questionable earthquakes

````python
#We plot and compare histograms of the intensities of tsunamis caused by earthquakes and volcanoes combinedly with the intensities of the tsunamis caused by questionable earthquakes
fig, axes = plt.subplots(nrows=1, ncols=2,
                        figsize=(10,10))

data[data['CAUSE_CODE']==4].hist(column='SOLOVIEV',ax=axes[0])# We plot the histogram for assessing tsunami intensities caused by volcanoes and earthquakes combinedly
data[data['CAUSE_CODE']==2].hist(column='SOLOVIEV',ax=axes[1])# We plot the histogram for assessing tsunami intensities caused by questionable earthquakes

# We label the y axis as 'Probability frequency' and the x axis as 'Tsunami intensity'
axes[0].set_ylabel('Probability frequency') 
axes[0].set_xlabel('Tsunami intensity') 
axes[1].set_ylabel('Probability frequency') 
axes[1].set_xlabel('Tsunami intensity')
````

#### Figure 8.(a) Histogram of intensity of tsunamis induced by volcanoes and earthquakes (left) (b) Histogram of intensity of tsunamis induced by questionable earthquakes (right)

The plots and figures shown above highlight the significant relationship between the intensity of tsunamis and their causes related to several gelogical hazards like earthquakes and volcanoes.

The stark contrast between figures 1 and 2 highlights that majority of the tsunamis are caused by unquestionable earthquakes after 2000.

Figure 3 explores the tsunami events occuring along the Ring of Fire and we observe that most of the tsunamis are concentrated in this region. This is attributed to the presence of subduction zones and mantle plume hotspots along the Pacific Rim which trigger volcanic eruptions and subsequently lead to earthquakes and ultimately tsunamis (Hinga, 2015). 

On comparison of the subplots in figure 4 as well as the histograms in figures 5 and 6, we can observe that relatively more number of high intensity tsunamis occur due to earthquakes than due to volcanoes.From figure 6, we notice that the highest tsunami intensity of 8 is attained due to an earthquake-induced tsunami.This can be linked to the larger and more frequent direct impacts of earthquake-induced seismic waves on the sea floor compared to the occasional volcanic eruptions triggered mostly by upwelling magma resulting in mantle plumes (hotspots). 


The histogram in figure 8(b) tells us that three tsunami events (two with intensities ranging from 1-1.25 and one of intensity between 2.75-3) are caused by questionable earthquakes which can be excluded from the tsunami events data represented in Figure 8(a) to improve the precision of our assessment. We can also infer that the negative intensity values along the x axis in the plots indicate that the runup height of the wave associated with those events was below the average runup height (from equation (1)).

There is no significant difference observed in figures 6 and 7, ther we extend our code to plot figures 8 (a) and 8 (b) and distinguish the intensities of tsunamis caused by volcanoes and earthquakes from those caused by questionable earthquakes  

Plot 6 shows that the maximum number of tsunamis (16) induced by volcanoes have an intensity between 0.5-1 while plot 7 portrays that the maximum number of earthquake-induced tsunamis (approximately 310) have an intensity around 0-1. This restrengthens the argument that there could be an inter-connectedness between the the tsunamis caused by volcanoes and earthquakes, whereby one geological event precedes the other.
